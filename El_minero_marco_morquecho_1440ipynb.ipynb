## EL minero


```python
# ======================================================
# ⛏️ CBTis 128
# 📚 Programación Orientada a Objetos 2026
# 🤖 Especialidad: Inteligencia Artificial
# 👨‍🏫 Docente: Eliseo Nava
# 👨‍🎓 Alumno: Marco Morquecho
# ⛏️ Proyecto: Sistema de Mineros
# ======================================================

# =========================
# ⛏️ Lista de mineros
# =========================
mineros = []

# =========================
# ➕ Agregar minero
# =========================
def agregarxxx():

    nombre = input("Ingresa el nombre del minero: ")

    mineros.append(nombre)

    print("Minero agregado correctamente")

# =========================
# 📖 Leer minero
# =========================
def leerxxx():

    posicion = int(input("Ingresa la posición: "))

    if posicion < len(mineros):
        print("Minero:", mineros[posicion])

    else:
        print("Posición no válida")

# =========================
# 📜 Listar mineros
# =========================
def listarxxxx():

    if len(mineros) == 0:
        print("No hay mineros registrados")

    else:

        i = 0

        while i < len(mineros):

            print(i, "-", mineros[i])

            i += 1

# =========================
# 🔍 Buscar minero
# =========================
def buscarxxx():

    buscar = input("Ingresa el nombre del minero: ")

    if buscar in mineros:
        print("Minero encontrado")

    else:
        print("Minero no encontrado")

# =========================
# ✏️ Modificar minero
# =========================
def modificarxxx():

    posicion = int(input("Ingresa la posición a modificar: "))

    if posicion < len(mineros):

        nuevo = input("Ingresa el nuevo nombre: ")

        mineros[posicion] = nuevo

        print("Minero modificado correctamente")

    else:
        print("Posición no válida")

# =========================
# ❌ Borrar minero
# =========================
def borrarxxx():

    borrar = input("Ingresa el nombre del minero a borrar: ")

    if borrar in mineros:

        mineros.remove(borrar)

        print("Minero eliminado")

    else:
        print("Minero no encontrado")

# =========================
# 📌 Menú principal
# =========================
while True:

    print("\n===== MENÚ MINEROS ⛏️ =====")
    print("1. Agregar minero")
    print("2. Leer minero")
    print("3. Listar mineros")
    print("4. Buscar minero")
    print("5. Modificar minero")
    print("6. Borrar minero")
    print("7. Salir")

    opcion = int(input("Selecciona una opción: "))

    if opcion == 1:
        agregarxxx()

    elif opcion == 2:
        leerxxx()

    elif opcion == 3:
        listarxxxx()

    elif opcion == 4:
        buscarxxx()

    elif opcion == 5:
        modificarxxx()

    elif opcion == 6:
        borrarxxx()

    elif opcion == 7:
        print("Programa finalizado ⛏️")
        break

    else:
        print("Opción no válida")
```
